```python
import pandas as pd

# Read column names for all-weeks-countries.tsv
countries_df = pd.read_csv('all-weeks-countries.tsv', sep='\t')

# Filter for United States
us_df = countries_df[countries_df['country_name'] == 'United States']

# Filter for Japan
japan_df = countries_df[countries_df['country_name'] == 'Japan']

# Filter for Germany
germany_df = countries_df[countries_df['country_name'] == 'Germany']

# Filter for Mexico
mexico_df = countries_df[countries_df['country_name'] == 'Mexico']

# Define column names
column_names = {
    'show_title': 'Show Title',
    'cumulative_weeks_in_top_10': 'Cumulative Weeks in Netflix Top 10'
}

# Export to Excel
us_df[us_df['category'] == 'TV'] \
    .sort_values(by='cumulative_weeks_in_top_10', ascending=False) \
    .drop_duplicates(subset='show_title') \
    [['cumulative_weeks_in_top_10', 'show_title']] \
    .rename(columns=column_names) \
    .head(20) \
    .to_excel('top_20_us_tv_shows.xlsx', index=False)

japan_df[japan_df['category'] == 'TV'] \
    .sort_values(by='cumulative_weeks_in_top_10', ascending=False) \
    .drop_duplicates(subset='show_title') \
    [['cumulative_weeks_in_top_10', 'show_title']] \
    .rename(columns=column_names) \
    .head(20) \
    .to_excel('top_20_japan_tv_shows.xlsx', index=False)

germany_df[germany_df['category'] == 'TV'] \
    .sort_values(by='cumulative_weeks_in_top_10', ascending=False) \
    .drop_duplicates(subset='show_title') \
    [['cumulative_weeks_in_top_10', 'show_title']] \
    .rename(columns=column_names) \
    .head(20) \
    .to_excel('top_20_germany_tv_shows.xlsx', index=False)

mexico_df[mexico_df['category'] == 'TV'] \
    .sort_values(by='cumulative_weeks_in_top_10', ascending=False) \
    .drop_duplicates(subset='show_title') \
    [['cumulative_weeks_in_top_10', 'show_title']] \
    .rename(columns=column_names) \
    .head(20) \
    .to_excel('top_20_mexico_tv_shows.xlsx', index=False)
```
